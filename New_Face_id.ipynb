```python
import cv2
import face_recognition
import torch
from matplotlib import pyplot as plt

# Load YOLOv5s model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Step 1: Upload your image to Colab
from google.colab import files
uploaded = files.upload()

# Step 2: Load uploaded image
import os
image_path = next(iter(uploaded))
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 3: Face Detection
face_locations = face_recognition.face_locations(image_rgb)
for top, right, bottom, left in face_locations:
    cv2.rectangle(image_rgb, (left, top), (right, bottom), (0, 255, 0), 2)
    cv2.putText(image_rgb, "Detected Face", (left, top - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

# Step 4: Object Detection (YOLO)
results = model(image)
for *box, conf, cls in results.xyxy[0]:
    x1, y1, x2, y2 = map(int, box)
    label = model.names[int(cls)]
    cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (255, 0, 255), 2)
    cv2.putText(image_rgb, label, (x1, y1 - 5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

# Step 5: Show final image
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis("off")
plt.title("Face + Object Recognition Result")
plt.show()

```

```python
import cv2
import torch
from matplotlib import pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Load image
image_path = next(iter(uploaded))
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Load YOLOv5 pretrained model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Run detection
results = model(image_rgb)

# Draw boxes
for *box, conf, cls in results.xyxy[0]:
    x1, y1, x2, y2 = map(int, box)
    label = model.names[int(cls)]
    cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.putText(image_rgb, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.8, (255, 255, 0), 2)

# Show result
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis("off")
plt.title("Object Detection (YOLOv5)")
plt.show()

```
